```python
def abre_chrome_retorna_driver():
    print('\nFunction --> abre_chrome_retorna_driver')

    """
    █▀█ █░█ ▀█▀
    █▄█ █▄█ ░█░
    """
    
    # driver  • <class 'selenium.webdriver.chrome.webdriver.WebDriver'>  ○ driver connection object, will be used to browser around
    
    print("abre_chrome_retorna_driver(): iniciando novo selenium")
    path_driver = r"chromedriver.exe" # neste caso, o binário do driver deve estar na mesma pasta que o script

    options = webdriver.ChromeOptions()
    options.add_experimental_option('useAutomationExtension', False)

    driver = webdriver.Chrome(executable_path = path_driver, options = options)
    driver.maximize_window()
    print("abre_chrome_retorna_driver(): retornando driver")
    print('Function <-- abre_chrome_retorna_driver\n')
    return driver
```

```python
def abre_automation_practice_confirma_pagina(driver, url):
    print('\nFunction --> abre_automation_practice_confirma_pagina')
    
    """   
    █ █▄░█
    █ █░▀█    
    """
        # driver  • <class 'selenium.webdriver.chrome.webdriver.WebDriver'>  ○ driver connection object, will be used to browser around
        # url     • <str>                                                    ○ url we will browse to  
        
    """
    █▀█ █░█ ▀█▀
    █▄█ █▄█ ░█░
    """
        # -1          ○ página carregou mas o heading possui outro texto diferente de <authentication>
        # -2          ○ site errado, fora do ar ou elemento de conferência que esperamos não existe mais
        # -3          ○ parâmetro inválido para url, precisa ser string
#         ~~~~~~
        # True        ○ sucesso, página que desejamos foi aberta e conferida
    
    # -------------------------------------------------------------
    # ------------------ parameters printing ----------------------
    # -------------------------------------------------------------
    print(f'abre_automation_practice_confirma_pagina(): |url|: {url}')
    
    # -------------------------------------------------------------
    # ------------------ parameters validation --------------------
    # -------------------------------------------------------------
    if not isinstance(url,str):
        print(f'abre_automation_practice_confirma_pagina(): parâmetro inválido para url, precisa ser string')
        print('Function <-- abre_automation_practice_confirma_pagina\n')
        return -3
    
    # -------------------------------------------------------------
    # ------------------ browse to aimed url ----------------------
    # -------------------------------------------------------------
    driver.get(url)
    print(f'abre_automation_practice_confirma_pagina(): aberto')
    
    # -------------------------------------------------------------
    # ------- checking if aimed element is loaded on page ---------
    # -------------------------------------------------------------
    timer = Timer(10) # limite de 10 seg p site carregar
    while timer.not_expired:
        try:
            heading_element = driver.find_element_by_class_name("page-heading") #usaria o webdriverwait só que pra lista dá timeout, tem que ser elemento único
            print(f'abre_automation_practice_confirma_pagina(): |heading_element_innerText|: {heading_element.get_attribute("innerText")}')
            if 'AUTHENTICATION' in heading_element.get_attribute("innerText"):
                print(f'abre_automation_practice_confirma_pagina(): confirmado, página carregou e o cabeçalho indica que estamos no local correto (página de autenticação)')
                # --------------
                # -- SUCESSO ---
                # --------------
                print('Function <-- abre_automation_practice_confirma_pagina\n')
                return True 
            else:
                print(f'abre_automation_practice_confirma_pagina(): página carregou mas o heading possui outro texto; link está redirecionando para uma página que não é a que desejamos')
                print('Function <-- abre_automation_practice_confirma_pagina\n')
                return -1
        except NoSuchElementException:
            print(f'abre_automation_practice_confirma_pagina(): não encontramos elemento, site não carregou, mas ainda estamos dentro do tempo, vamos buscá-lo novamente')
            time.sleep(1) # p nao floodar os logs
            continue
            
    if timer.expired:
        print(f'abre_automation_practice_confirma_pagina(): tempo expirado, buscamos o elemento mas não encontramos, site errado, fora do ar ou elemento não existe mais')
        print('Function <-- abre_automation_practice_confirma_pagina\n')
        return -2
```

```python
def loga_site_automation(driver, email, senha):
    print('\nFunction --> loga_site_automation')
    
    """   
    █ █▄░█
    █ █░▀█    
    """
        # driver  • <class 'selenium.webdriver.chrome.webdriver.WebDriver'>  ○ driver connection object, will be used to browser around
        # email   • <str>                                                    ○ email to authenticate  
        # senha   • <str>                                                    ○ password to authenticate 
        
    """
    █▀█ █░█ ▀█▀
    █▄█ █▄█ ░█░
    """
        # -1          ○ parâmetro inválido para senha, precisa ser string
        # -2          ○ parâmetro inválido para email, precisa ser string
        # -3          ○ parâmetro inválido para email, e-mail sem @ 
        # -4          ○ campo de e-mail não disponível. impossível logar
        # -5          ○ campo de senha não disponível. impossível logar
        # -6          ○ botão Sign in p autenticar não foi encontrado
        # -7          ○ mensagem <Authentication failed>
        # -8          ○ nenhum elemento de alerta apareceu mas também não foi possível confirmar logon com sucesso
        # -9          ○ erro desconhecido, não foi possível logar
        # -10         ○ senha incorreta
#        ~~~~~~ 
        # True        ○ logon confirmado com sucesso
        
        
    # -------------------------------------------------------------
    # ------------------ parameters printing ----------------------
    # -------------------------------------------------------------
    print(f'loga_site_automation(): |email|: {email}')
    print(f'loga_site_automation(): |senha|: {senha}')
    
    # -------------------------------------------------------------
    # ------------------ parameters validation --------------------
    # -------------------------------------------------------------
    if not isinstance(senha,str) :
        print(f'loga_site_automation(): parâmetro inválido para senha, precisa ser string')
        print('Function <-- loga_site_automation\n') 
        return -1
    
    if not isinstance(email,str):
        print(f'loga_site_automation(): parâmetro inválido para email, precisa ser string')
        print('Function <-- loga_site_automation\n') 
        return -2
        
    if not '@' in email:
        print(f'loga_site_automation(): parâmetro inválido para email, como pode um e-mail sem @ ?')
        print('Function <-- loga_site_automation\n') 
        return -3
    
    # -------------------------------------------------------------
    # ----------------- typing e-mail on login field --------------
    # -------------------------------------------------------------
    print(f'loga_site_automation(): digitando e-mail no campo de login')
    try:
        campo_email = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "email")))
        campo_email.send_keys(email)
        
    except TimeoutException:
        print(f'loga_site_automation(): ok, não achamos o campo de e-mail pra digitar nele. impossível logar')
        print('Function <-- loga_site_automation\n')
        return -4
    
    # -------------------------------------------------------------
    # ---------------- typing password on login field -------------
    # -------------------------------------------------------------
    print(f'loga_site_automation(): digitando senha no campo de login')
    try:
        campo_senha = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "passwd")))
        campo_senha.send_keys(senha)
    except TimeoutException:
        print(f'loga_site_automation(): ok, não achamos o campo de senha pra digitar nele. impossível logar')
        print('Function <-- loga_site_automation')
        return -5
    
    # -------------------------------------------------------------
    # ----------------- click on <Sign in> button -----------------
    # -------------------------------------------------------------
    print(f'loga_site_automation(): procurando botão sign in para clicar e tentar autenticar no site')
    try:
        botao_sign_in = driver.find_element_by_id('SubmitLogin')
        botao_sign_in.click()
        print(f'loga_site_automation(): clicamos com sucesso no botão')
    except NoSuchElementException:
        print(f'loga_site_automation(): achamos os campos de login e senha mas o botão Sign in não foi encontrado. impossível tentar logar')
        print('Function <-- loga_site_automation\n')
        return -6
    
    # ------------------------------------------------------------
    # -------- check if some alert has showed up -----------------
    # ------------------------------------------------------------
    timer = Timer(10)
    while timer.not_expired:
        print(f'loga_site_automation(): tentando buscar se tem alerta no elemento')
        try:
            # -----------------------------------------------------------------------
            # atenção, este elemento não funciona com webdriver wait pois é uma lista
            # - com find elements by também está buscando lista vazia mesmo em loop -
            # ---------- mantenha o método execute_script pra este campo ------------
            # -----------------------------------------------------------------------
            qtd_msg_alerta = driver.execute_script('return document.getElementsByClassName("alert alert-danger").length')
            print(f'loga_site_automation(): |qtd_msg_alerta| :{qtd_msg_alerta}')
            
            for mensagem in range(qtd_msg_alerta):
                text_msg_alerta = driver.execute_script(f'return document.getElementsByClassName("alert alert-danger")[{mensagem}].innerText')
                print(f'loga_site_automation(): |text_msg_alerta|: ->{text_msg_alerta}<-')
                
                if 'Authentication failed' in text_msg_alerta:
                    print(f'loga_site_automation(): mensagem <Authentication failed>, usuário não existe ou senha está incorreta')
                    print('Function <-- loga_site_automation\n')
                    return -7
                
                if 'Invalid password' in text_msg_alerta:
                    print(f'loga_site_automation(): senha incorreta')
                    print('Function <-- loga_site_automation\n')
                    return -10
                
            break # se chegou aqui é pq achou elemento, não precisa ficar buscando ele sem parar até estourar o tempo
            
        except NoSuchElementException:
            print(f'loga_site_automation(): exception ao esperar elemento de alerta aparecer, vamos continuar buscado')
            continue
            
    # -------------------------------------------------------------
    # ------------- checking if logon was successed ---------------
    # -------------------------------------------------------------
    timer = Timer(10) # limite de 10 seg p site carregar criação de nova conta
    while timer.not_expired:
        try:
            heading_element = driver.find_element_by_class_name("page-heading") #usaria o webdriverwait só que pra lista dá timeout, tem que ser elemento único
            print(f'loga_site_automation(): |heading_element_innerText|: {heading_element.get_attribute("innerText")}')
            if 'MY ACCOUNT' in heading_element.get_attribute("innerText"):
                print(f'loga_site_automation(): confirmado, LOGIN FOI REALIZADO!')
                # --------------
                # -- SUCESSO ---
                # --------------
                print('Function <-- loga_site_automation\n')
                return True 
            else:
                print(f'loga_site_automation(): nenhum elemento de alerta apareceu mas também não foi possível confirmar logon com sucesso')
                print('Function <-- loga_site_automation\n')
                return -8
        except NoSuchElementException:
            print(f'loga_site_automation(): não encontramos elemento, site não carregou, mas ainda estamos dentro do tempo, vamos buscá-lo novamente')
            continue
    if timer.expired:       
        print(f'loga_site_automation(): erro desconhecido, não foi possível logar!')
        print('Function <-- loga_site_automation\n')
        return -9
```

```python
def seleciona_criar_conta(driver, dict_informacoes):
    print('\nFunction --> seleciona_criar_conta')
    
    """   
    █ █▄░█
    █ █░▀█    
    """
        # driver        • <class 'selenium.webdriver.chrome.webdriver.WebDriver'>  ○ driver connection object, will be used to browser around
        # dict_informacoes • <dict>                                             ○ informations that will be used to fill the form
        
    """
    █▀█ █░█ ▀█▀
    █▄█ █▄█ ░█░
    """
        # -1          ○ algum dos campos obrigatórios está em branco no dicionário recebido, cadastro dará erro
        # -2          ○ nao foi possivel digitar email para iniciar criação de conta
        # -3          ○ botao para selecionar <criar conta> não foi encontrado
        # -4          ○ campo que confere abertura de fomulário de cadastro não foi encontrado, formulário não abriu?
        # -5          ○ genero inválido passado por parâmetro, formulário só permite F ou M (feminino e masculino)
#          ~~~~~~~~~~
        # True        ○ sucesso, página que desejamos foi aberta e conferida
        
    # -------------------------------------------------------------
    # ------------------ parameters printing ----------------------
    # -------------------------------------------------------------
    print(f'seleciona_criar_conta(): |dict_informacoes|: {dict_informacoes}')
    
    # -------------------------------------------------------------
    # ------------------ parameters validation --------------------
    # -------------------------------------------------------------
    print(f'seleciona_criar_conta(): as keys que precisam ter valores obrigatórios são primeiro_nome, ultimo_nome, e-mail, senha, endereo, cidade, estado, codigo_postal, telefone, apelido_endereco')
    keys_obrigatorias = ['genero','primeiro_nome', 'ultimo_nome', 'e-mail', 'senha', 'endereco', 'cidade', 'estado', 'codigo_postal', 'telefone', 'apelido_endereco']
    
    for campo, valor in dict_informacoes.items():
        if campo in keys_obrigatorias and valor == "":
            # -------------------------------
            # campo obrigatório está vazio
            # -------------------------------
            print(f'seleciona_criar_conta(): no dicionário recebido por parâmetro, a chave {campo} está sem valor, se este campo não for preenchido, não há como completar o cadastro.')
            print('Function <-- seleciona_criar_conta\n')
            return -1
    if dict_informacoes['genero'] not in ['F','M']:
        print(f'seleciona_criar_conta(): atenção, gênero inválido, este formulário só permite selecionar senhor ou senhora, i.e masculino ou feminino; escolha com o qual você se identifica melhor para o cadastro')
        print('Function <-- seleciona_criar_conta\n')
        return -5
        
    # -------------------------------------------------------------
    # ------- filling email field to create an account ------------
    # -------------------------------------------------------------
    print(f'seleciona_criar_conta(): procurando campo pra digitar email pra criar conta')
    try:
        campo_email = driver.find_element_by_id('email_create')
        campo_email.send_keys(dict_informacoes['e-mail'])
        print(f'seleciona_criar_conta(): digitamos o e-mail')
    except NoSuchElementException:
        print(f'seleciona_criar_conta(): nao foi possivel digitar email para iniciar criação de conta')
        print('Function <-- seleciona_criar_conta\n')
        return -2
    
    # -------------------------------------------------------------
    # ---------------- clicking on <CREATE ACCOUNT> ---------------
    # -------------------------------------------------------------
    print(f'seleciona_criar_conta(): procurando botao Create Account')
    try:
        botao_create_account = driver.find_element_by_id('SubmitCreate')
        botao_create_account.click()
        print(f'seleciona_criar_conta(): clicamos com sucesso no botão')
    except NoSuchElementException:
        print(f'seleciona_criar_conta(): impossível achar botao para selecionar <criar conta>, mudou id?')
        print('Function <-- seleciona_criar_conta\n')
        return -3
    
    # -------------------------------------------------------------
    # --------- checking if expected page is loaded ---------------
    # -------------------------------------------------------------
    print(f'seleciona_criar_conta(): conferindo se campo de primeiro nome está carregado para inciar preenchimento do formulário de cadastro')
    try: 
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "customer_firstname")))
        print(f'seleciona_criar_conta(): ok, confirmado estamos na página com formulário de cadastro')
        # --------------
        # -- SUCESSO ---
        # --------------
        print('Function <-- seleciona_criar_conta\n')
        return True
    except TimeoutException:
        
        print(f' então vamos verificar se apareceu mensagem de erro')
        print(f'seleciona_criar_conta(): vamos buscar o elemento de alerta')
        try:
            elemento_alerta = driver.find_element_by_id("create_account_error")
            texto_elemento_alerta = elemento_alerta.get_attribute("innerText")
            print(f'seleciona_criar_conta(): |texto_elemento_alerta|: {texto_elemento_alerta}')
            if 'An account using this email address has already been registered' in texto_elemento_alerta:
                print(f'seleciona_criar_conta():  já tem uma conta com este endereço de e-mail, por favor escolha outro email')
                print('Function <-- seleciona_criar_conta\n')
                return -4
            elif texto_elemento_alerta != "": #tem alguma mensagem ainda nao mapeada
                print(f'seleciona_criar_conta(): atenção mensagem não mapeada no elemento de erro após tentar criação de nova conta : {texto_elemento_alerta}')
                print('Function <-- seleciona_criar_conta\n')
                return -5

        except NoSuchElementException:
            print(f'clica_register_verifica_mensagem(): não apareceu mensagem no elemento de alerta e não encontramos o campo pra inserir primeiro nome do cliente; página de cadastro não carregou ou ID deste elemento mudou')
            print('Function <-- seleciona_criar_conta\n')
            return -6
```

```python
class Timer:
    def __init__(self, duration=10):
        self.duration = float(duration)
        self.start = time.perf_counter()
        # print("The timer has started. Self.start: " + str(self.start))

    def reset(self):
        self.start = time.perf_counter()
        # print("The timer has been reset. Self.start: " + str(self.start))

    def explode(self):
        self.duration = 0
        # print("The timer has been force-expired.")

    def increment(self, increment=0):
        self.duration += increment
        # print("The timer has been incremented by " + str(increment) + " seconds")

    @property
    def not_expired(self):
        # duration == -1 means dev wants a infinite loop/Timer
        if self.duration == -1:
            return True
        return False if time.perf_counter() - self.start > self.duration else True

    @property
    def expired(self):
        return not self.not_expired

    @property
    def at(self):
        # print("The timer is running. Self.at: " + str(time.perf_counter() - self.start))
        return time.perf_counter() - self.start
```
